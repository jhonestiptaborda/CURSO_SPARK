# Introducción a PySpark

## Tarea 1: Creación de una SparkSession

En esta práctica se creará una sesión local de Apache Spark y se validarán sus principales propiedades.

````python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .master("local[*]")
    .appName("Curso Spark Introducción")
    .getOrCreate()
)
````

````python
print("Versión de Spark:", spark.version)
print("Versión de Python:", spark.sparkContext.pythonVer)
print("Nombre de la aplicación:", spark.sparkContext.appName)
print("Master:", spark.sparkContext.master)
````

## Creación del primer DataFrame

En esta práctica se crea un DataFrame a partir de un archivo JSON, siguiendo el ejemplo de la documentación oficial de Spark SQL.

````python
df_people = spark.read.json("../data/raw/people.json")
df_people.head(5)
````

````python
df_people.show()
````

## Inspección del esquema del DataFrame

La documentación oficial de Spark utiliza el método `printSchema()` para mostrar la estructura del DataFrame en forma de árbol.

Este método permite conocer:

- Los nombres de las columnas.
- Los tipos de datos detectados por Spark.
- Si una columna permite valores nulos.

El comando utilizado es:

```python
df_people.printSchema()

````python
df_people.printSchema()
````

## Selección de columnas en un DataFrame

La documentación oficial de Spark muestra que un DataFrame permite seleccionar una o varias columnas mediante el método `select()`.

En este ejemplo se seleccionará únicamente la columna `name`.

El código utilizado es:

```python
df_people.select("name").show()

````python
df_people.select("name").show()
````

## Selecionar todo e incrementar la edad en 1

````python
df_people.select(
    df_people["name"],
    df_people["age"],
    (df_people["age"] + 1).alias("age_plus_one")
).show()
````

## Selecionar la gente que es mayor de 21

````python
df_people.filter(df_people["age"] > 21).show()
````
